## Visualization & Explanatory Analysis

Tahap ini dilakukan untuk meningkatkan pemahaman tentang data, menyampaikan informasi dengan visualisasi yang memungkinkan untuk lebih dimengerti, dan mendukung pengambilan keputusan berbasis data.

### Pertanyaan 1:

Apakah ada hubungan yang saling berpengaruh antara kelembaban udara (DEWP), suhu (TEMP), dan polutan seperti PM2.5 dan PM10?

```python
# Menghitung korelasi antara kelembaban (DEWP), suhu (TEMP), PM2.5, dan PM10
correlation_matrix = AirPolution_df[['DEWP', 'TEMP', 'PM2.5', 'PM10']].corr()
print("Korelasi antara kelembaban, suhu, dan level polutan:")
print(correlation_matrix)
```

Heatmap ini menunjukkan korelasi antar faktor-faktor yang mempengaruhi kualitas udara

```python
# Visualisasikan heatmap untuk korelasi
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriks Korelasi antara Kelembaban, Suhu, PM2.5, dan PM10')
plt.show()
```

### Pertanyaan 2:

Bagaimana perbandingan konsentrasi polutan di stasiun-stasiun? Stasiun apa yang mempunyai kualitas udara terbaik dan terburuk?

```python
# Menghitung rata-rata konsentrasi polutan untuk setiap stasiun
Polutan_columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'O3']
average_Polutan = AirPolution_df.groupby('station')[Polutan_columns].mean().reset_index()
```

```python
import pandas as pd
from datetime import datetime

# Mengonversi kolom 'datetime' menjadi tipe datetime
AirPolution_df['datetime'] = pd.to_datetime(AirPolution_df['datetime'])

# Menghitung Recency
last_date = AirPolution_df['datetime'].max()
AirPolution_df['Recency'] = (last_date - AirPolution_df['datetime']).dt.days

# Menghitung Frequency
frequency_df = AirPolution_df[AirPolution_df['SO2'] > 100].groupby('station').agg({
    'Recency': 'min',  # Ambil Recency terkecil
    'datetime': 'count',  # Hitung jumlah hari polusi tinggi
    'SO2': 'sum'  # Total SO2
}).reset_index()

frequency_df.rename(columns={'datetime': 'Frequency', 'SO2': 'Monetary'}, inplace=True)

# Menampilkan RFM analysis
print(frequency_df)
```

```python
# Mengonversi kolom 'datetime' menjadi tipe datetime
AirPolution_df['datetime'] = pd.to_datetime(AirPolution_df['datetime'])

# Menghitung rata-rata untuk setiap polutan di setiap stasiun
geo_df = AirPolution_df.groupby('station')[['SO2', 'PM2.5', 'PM10', 'NO2', 'O3', 'CO']].mean().reset_index()

# Menentukan batas bin dan label untuk SO2
bins_so2 = [0, 50, 100, float('inf')]
labels_so2 = ['Rendah', 'Sedang', 'Tinggi']

# Mengelompokkan data SO2 ke dalam bin
geo_df['Kategori_SO2'] = pd.cut(geo_df['SO2'], bins=bins_so2, labels=labels_so2)

# Menentukan batas bin dan label untuk PM2.5
bins_pm25 = [0, 12, 35.4, float('inf')]
labels_pm25 = ['Rendah', 'Sedang', 'Tinggi']

# Mengelompokkan data PM2.5 ke dalam bin
geo_df['Kategori_PM2.5'] = pd.cut(geo_df['PM2.5'], bins=bins_pm25, labels=labels_pm25)

# Menentukan batas bin dan label untuk PM10
bins_pm10 = [0, 50, 150, float('inf')]
labels_pm10 = ['Rendah', 'Sedang', 'Tinggi']

# Mengelompokkan data PM10 ke dalam bin
geo_df['Kategori_PM10'] = pd.cut(geo_df['PM10'], bins=bins_pm10, labels=labels_pm10)

# Menentukan batas bin dan label untuk NO2
bins_no2 = [0, 40, 80, float('inf')]
labels_no2 = ['Rendah', 'Sedang', 'Tinggi']

# Mengelompokkan data NO2 ke dalam bin
geo_df['Kategori_NO2'] = pd.cut(geo_df['NO2'], bins=bins_no2, labels=labels_no2)

# Menentukan batas bin dan label untuk O3
bins_o3 = [0, 50, 100, float('inf')]
labels_o3 = ['Rendah', 'Sedang', 'Tinggi']

# Mengelompokkan data O3 ke dalam bin
geo_df['Kategori_O3'] = pd.cut(geo_df['O3'], bins=bins_o3, labels=labels_o3)

# Menentukan batas bin dan label untuk CO
bins_co = [0, 4, 9, float('inf')]
labels_co = ['Rendah', 'Sedang', 'Tinggi']

# Mengelompokkan data CO ke dalam bin
geo_df['Kategori_CO'] = pd.cut(geo_df['CO'], bins=bins_co, labels=labels_co)

# Menghitung jumlah stasiun dalam setiap kategori untuk semua polutan
category_counts_so2 = geo_df['Kategori_SO2'].value_counts().reset_index()
category_counts_so2.columns = ['Kategori_SO2', 'Jumlah Stasiun']

category_counts_pm25 = geo_df['Kategori_PM2.5'].value_counts().reset_index()
category_counts_pm25.columns = ['Kategori_PM2.5', 'Jumlah Stasiun']

category_counts_pm10 = geo_df['Kategori_PM10'].value_counts().reset_index()
category_counts_pm10.columns = ['Kategori_PM10', 'Jumlah Stasiun']

category_counts_no2 = geo_df['Kategori_NO2'].value_counts().reset_index()
category_counts_no2.columns = ['Kategori_NO2', 'Jumlah Stasiun']

category_counts_o3 = geo_df['Kategori_O3'].value_counts().reset_index()
category_counts_o3.columns = ['Kategori_O3', 'Jumlah Stasiun']

category_counts_co = geo_df['Kategori_CO'].value_counts().reset_index()
category_counts_co.columns = ['Kategori_CO', 'Jumlah Stasiun']

# Visualisasi kategori
plt.figure(figsize=(20, 12))

# SO2
plt.subplot(2, 3, 1)
sns.barplot(x='Kategori_SO2', y='Jumlah Stasiun', data=category_counts_so2, palette='Blues_d')
plt.title('Jumlah Stasiun Berdasarkan Kategori SO2')
plt.xlabel('Kategori SO2')
plt.ylabel('Jumlah Stasiun')

# PM2.5
plt.subplot(2, 3, 2)
sns.barplot(x='Kategori_PM2.5', y='Jumlah Stasiun', data=category_counts_pm25, palette='Greens_d')
plt.title('Jumlah Stasiun Berdasarkan Kategori PM2.5')
plt.xlabel('Kategori PM2.5')
plt.ylabel('Jumlah Stasiun')

# PM10
plt.subplot(2, 3, 3)
sns.barplot(x='Kategori_PM10', y='Jumlah Stasiun', data=category_counts_pm10, palette='Reds_d')
plt.title('Jumlah Stasiun Berdasarkan Kategori PM10')
plt.xlabel('Kategori PM10')
plt.ylabel('Jumlah Stasiun')

# NO2
plt.subplot(2, 3, 4)
sns.barplot(x='Kategori_NO2', y='Jumlah Stasiun', data=category_counts_no2, palette='Purples_d')
plt.title('Jumlah Stasiun Berdasarkan Kategori NO2')
plt.xlabel('Kategori NO2')
plt.ylabel('Jumlah Stasiun')

# O3
plt.subplot(2, 3, 5)
sns.barplot(x='Kategori_O3', y='Jumlah Stasiun', data=category_counts_o3, palette='Oranges_d')
plt.title('Jumlah Stasiun Berdasarkan Kategori O3')
plt.xlabel('Kategori O3')
plt.ylabel('Jumlah Stasiun')

# CO
plt.subplot(2, 3, 6)
sns.barplot(x='Kategori_CO', y='Jumlah Stasiun', data=category_counts_co, palette='Greys_d')
plt.title('Jumlah Stasiun Berdasarkan Kategori CO')
plt.xlabel('Kategori CO')
plt.ylabel('Jumlah Stasiun')

plt.tight_layout()
plt.show()

# Menampilkan hasil pengelompokan
print(geo_df)
```
